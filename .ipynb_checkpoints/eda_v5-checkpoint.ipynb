```python
#Comuneros_Primavera [1]
discharges_Comuneros_Primavera = pd.read_csv('data/DF_Descargas_Comuneros_Primavera_2018_2019.csv', header=0, delimiter=',', index_col=0,
                         names=['date','longitude','latitude','polarity','magnitude','current'], parse_dates=['date'])
outages_Comuneros_Primavera = pd.read_csv('data/Fallas_Comuneros_Primavera_2010_2019.csv', header=0, delimiter=',', index_col=0,
                     names=['date','year','time','cause','outages_number','r_inf','r_sup'], parse_dates=['date'])
towers_Comuneros_Primavera = pd.read_csv('data/Georef_torres_Comuneros_Primavera.csv', header=0, delimiter=',', names=['longitude','latitude'])
```

```python
#Cerromatoso_Primavera [2]
discharges_Cerromatoso_Primavera = pd.read_csv('data/DF_Descargas_Cerromatoso_Primavera_2007_2020.csv', header=0, delimiter=',', index_col=0,
                         names=['date','longitude','latitude','polarity','magnitude','current'], parse_dates=['date'])
outages_Cerromatoso_Primavera = pd.read_csv('data/Fallas_Cerro_Primavera_2010_2019.csv', header=0, delimiter=',', index_col=0,
                     names=['date','year','time','cause','outages_number','r_inf','r_sup'], parse_dates=['date'])
towers_Cerromatoso_Primavera = pd.read_csv('data/Georef_torres_Primavera_Cerromatoso.csv', header=0, delimiter=',', names=['longitude','latitude'])
```

```python
discharges_Cerromatoso_Primavera["linea"] = 2
outages_Cerromatoso_Primavera["linea"] = 2
towers_Cerromatoso_Primavera["linea"] = 2
```

```python
#Virginia_SanCarlos [3]
discharges_Virginia_SanCarlos = pd.read_csv('data/DF_Descargas_LaVirginia_SanCarlos_2007_2020.csv', header=0, delimiter=',', index_col=0,
                         names=['date','longitude','latitude','polarity','magnitude','current'], parse_dates=['date'])
outages_Virginia_SanCarlos = pd.read_csv('data/Fallas_Virginia_SanCarlos_2010_2019.csv', header=0, delimiter=',', index_col=0,
                     names=['date','year','time','cause','outages_number','r_inf','r_sup'], parse_dates=['date'])
towers_Virginia_SanCarlos = pd.read_csv('data/Georef_torres_San Carlos_Virginia.csv', header=0, delimiter=',', names=['longitude','latitude'])
```

```python
discharges_Virginia_SanCarlos["linea"] = 3
outages_Virginia_SanCarlos["linea"] = 3
towers_Virginia_SanCarlos["linea"] = 3
```

```python
discharges = pd.concat([discharges_Comuneros_Primavera, discharges_Cerromatoso_Primavera, discharges_Virginia_SanCarlos], ignore_index=True)
outages = pd.concat([outages_Comuneros_Primavera, outages_Cerromatoso_Primavera, outages_Virginia_SanCarlos], ignore_index=True)
towers = pd.concat([towers_Comuneros_Primavera, towers_Cerromatoso_Primavera, towers_Virginia_SanCarlos], ignore_index=True)
```

```python
outages['date'] = pd.to_datetime(outages['date'])

def previous_discharges_by_time(outage_date, time_i, time_f):
    datetime_f = outage_date - timedelta(minutes=time_f)
    datetime_i = datetime_f - timedelta(minutes=time_i)
    previous_discharges = discharges[(discharges['date']>datetime_i) &
                        (discharges['date']<datetime_f)]
    return previous_discharges

for outage_date in outages['date']:
    previous_discharges = previous_discharges_by_time(outage_date=outage_date, time_i=25, time_f=5)
    print(previous_discharges)
```
